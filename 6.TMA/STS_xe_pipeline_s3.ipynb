```python
samples = ['XM-6','T640T1','T1088','T1982','KHE007','T1489','T943','T554','T1821','T2008','KHE008','T1739','T487','T1706','T1768','T1879','T924','T1440','T143','T701','T516','KHE004','T876','T1668','T640T2','T877','T937','T1057','T976']
```

```python
for sample in samples:
    adata_valid = sc.read(f'/cluster3/yflu/STS/TMA/separated/{sample}_counts.h5ad')
    new_model = models.Model.load('/cluster3/yflu/STS/celltypist/STS_pega_interact_full_40_5k.pkl')
    adata_1 = sc.read(f'/cluster3/yflu/STS/TMA/separated/{sample}/{sample}_preprocessed.h5ad')
    adata_1.obs.index = adata_1.obs.cell_id
    adata_valid.obs.index = adata_1.obs.cell_id
    adata_valid.obs["louvain_labels"] = adata_1.obs["louvain_labels"]
    sc.pp.normalize_total(adata_valid, target_sum=1e4)
    sc.pp.log1p(adata_valid)
    predictions = celltypist.annotate(adata_valid, model = new_model, majority_voting = True,over_clustering = "louvain_labels")
    adata_predictions = predictions.to_adata()
    adata_1.obs["predicted_labels"] = adata_valid.obs["predicted_labels"]
    adata_1.obs["majority_voting"] = adata_valid.obs["majority_voting"]
    adata_1.obs["predicted_labels"].value_counts()
    adata_1.write(f"/cluster3/yflu/STS/TMA/separated/{sample}/{sample}_celltypist.h5ad")
    sc.pl.umap(adata_1,color = ["predicted_labels","majority_voting"], show=False)
    plt.savefig(f"/cluster3/yflu/STS/TMA/separated/{sample}/umap_celltypist_5K.pdf", format="pdf", bbox_inches="tight")
    sc.pl.spatial(adata_1,color = ["predicted_labels","majority_voting"],spot_size=10, show=False)
    plt.savefig(f"/cluster3/yflu/STS/TMA/separated/{sample}/spatial_celltypist_5K.pdf", format="pdf", bbox_inches="tight")
```
