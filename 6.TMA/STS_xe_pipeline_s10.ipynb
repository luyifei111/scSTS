```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from tqdm import tqdm
import scanpy as sc

def add_disease_scores_and_plot_density_scatter_per_sample(
    adata, 
    disease_to_geneset, 
    score_key="score_mtx", 
    tumor_celltype="Tumor cells",
    save=None,
    figsize=(12,12),
    cmap_density="viridis",
    point_size=10,
    mode="density"  # "density"=普通密度, "weighted"=表达量加权密度
):
    """
    按 Sample 分批计算 Tumor cells 的局部密度，并绘制密度散点图。
    
    Parameters
    ----------
    adata : AnnData
    disease_to_geneset : dict
        疾病 -> 基因集名称
    score_key : str
        存放评分矩阵的 obsm 键名
    tumor_celltype : str
        仅提取该类型的细胞
    save : str or None
        保存路径
    figsize : tuple
        图像大小
    cmap_density : str
        colormap
    point_size : int
        点大小
    mode : str
        "density" -> 普通密度
        "weighted" -> 表达量加权密度
    """
    # ================================
    # 1. 提取 Tumor cells 的基因集评分
    # ================================
    expr = np.full(adata.n_obs, np.nan, dtype=float)
    score_mtx = adata.obsm[score_key]
    
    for sample in tqdm(adata.obs["Sample"].unique(), desc="Processing samples"):
        disease = adata.obs.loc[adata.obs["Sample"] == sample, "Disease"].unique()[0]
        geneset = disease_to_geneset.get(disease, None)
        if geneset is None or geneset not in score_mtx.columns:
            continue
        
        mask = (adata.obs["Sample"] == sample) & (adata.obs["celltype"] == tumor_celltype)
        if np.sum(mask) == 0:
            continue
        
        expr_sample = score_mtx.loc[adata.obs_names[mask], geneset].values
        expr[mask] = expr_sample
    
    # ================================
    # 2. scale
    # ================================
    expr_scaled = np.full_like(expr, np.nan, dtype=float)
    valid_mask = ~np.isnan(expr)
    expr_scaled[valid_mask] = StandardScaler().fit_transform(expr[valid_mask].reshape(-1,1)).flatten()
    
    adata.obs["matched_marker_expr_scaled"] = expr_scaled
    print("Scaled expression range (Tumor cells only):",
          np.nanmin(expr_scaled), np.nanmax(expr_scaled))
    
    # ================================
    # 3. 按 Sample 计算密度
    # ================================
    coords = adata.obsm["spatial"]
    densities = np.full(adata.n_obs, np.nan)
    
    for sample in tqdm(adata.obs["Sample"].unique(), desc="Calculating densities per sample"):
        mask = (adata.obs["Sample"] == sample) & (adata.obs["celltype"] == tumor_celltype)
        if np.sum(mask) == 0:
            continue
        coords_sample = coords[mask]
        kde = gaussian_kde(coords_sample.T)
        
        if mode == "density":
            densities[mask] = kde(coords_sample.T)
        elif mode == "weighted":
            # 表达量加权密度（确保非负）
            expr_sample = expr_scaled[mask]
            expr_shifted = expr_sample - np.min(expr_sample)
            densities[mask] = kde(coords_sample.T) * expr_shifted
        else:
            raise ValueError("mode 必须是 'density' 或 'weighted'")
    
    # ================================
    # 4. 绘图
    # ================================
    valid_mask = ~np.isnan(densities)
    # 按密度排序，保证低密度点先画
    idx = densities[valid_mask].argsort()
    scatter_coords = coords[valid_mask][idx]
    scatter_density = densities[valid_mask][idx]
    
    plt.figure(figsize=figsize)
    plt.scatter(
        scatter_coords[:,0],
        scatter_coords[:,1],
        c=scatter_density,
        s=point_size,
        cmap=cmap_density,
        alpha=0.9,
        edgecolors="none"
    )
    plt.gca().invert_yaxis()
    plt.axis("off")
    plt.colorbar(label="Local density")
    
    if save is not None:
        plt.savefig(save, format="png", bbox_inches="tight", dpi=600)
        print(f"✅ 密度散点图已保存: {save}")
    else:
        plt.show()
    
    return adata
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

def plot_multi_gene_density(
    adata, 
    genes, 
    score_key="score_mtx", 
    tumor_label="Tumor cells",
    mode="weighted",   # "density"=普通密度, "weighted"=表达量加权密度
    point_size=8,
    cmap="viridis",
    ncols=2,
    figsize=(12, 5),
    bg_size=5,
    bg_alpha=0.3,
    share_scale=False,  # 是否共享颜色范围
    save=None
):
    """
    绘制多个基因的密度点图 (子图形式)
    - tumor_label 对应的细胞参与 KDE 并着色
    - 其他细胞标灰色背景
    """

    if score_key not in adata.obsm:
        raise KeyError(f"{score_key} not found in adata.obsm")

    score_mtx = adata.obsm[score_key]

    n_genes = len(genes)
    nrows = int(np.ceil(n_genes / ncols))

    # figure，留出右侧空间放 colorbar
    fig, axes = plt.subplots(
        nrows, ncols, 
        figsize=(figsize[0], figsize[1] * nrows),
        gridspec_kw={"wspace": 0.2, "hspace": 0.3, "right": 0.85}
    )
    axes = np.array(axes).reshape(-1)

    # 背景数据（非肿瘤）
    mask_tumor = adata.obs["celltype"] == tumor_label
    mask_other = ~mask_tumor
    coords_other = np.asarray(adata.obsm["spatial"][mask_other], dtype=float)

    all_values = []
    densities_dict = {}

    # 先计算所有基因的密度，便于共享 scale
    for gene in genes:
        if gene not in score_mtx.columns:
            continue
        coords_tumor = np.asarray(adata.obsm["spatial"][mask_tumor], dtype=float)
        expr_tumor = np.asarray(score_mtx.loc[adata.obs_names[mask_tumor], gene].values, dtype=float)

        kde = gaussian_kde(coords_tumor.T)
        densities = kde(coords_tumor.T)

        if mode == "weighted":
            expr_shifted = expr_tumor - np.nanmin(expr_tumor)
            densities = densities * expr_shifted

        densities_dict[gene] = (coords_tumor, densities)
        all_values.append(densities)

    # 颜色范围
    vmin = min(np.min(v) for v in all_values) if share_scale else None
    vmax = max(np.max(v) for v in all_values) if share_scale else None

    sc = None
    for i, gene in enumerate(genes):
        ax = axes[i]

        if gene not in densities_dict:
            ax.set_title(f"{gene} (not found)")
            ax.axis("off")
            continue

        coords_tumor, densities = densities_dict[gene]

        # 排序（避免高值被遮盖）
        idx = densities.argsort()
        coords_tumor = coords_tumor[idx]
        densities = densities[idx]

        # 背景
        ax.scatter(
            coords_other[:,0],
            coords_other[:,1],
            c="lightgray",
            s=bg_size,
            alpha=bg_alpha,
            edgecolors="none"
        )

        # 肿瘤细胞
        sc = ax.scatter(
            coords_tumor[:,0],
            coords_tumor[:,1],
            c=densities,
            s=point_size,
            cmap=cmap,
            alpha=0.9,
            edgecolors="none",
            vmin=vmin,
            vmax=vmax
        )

        ax.set_title(gene, fontsize=12)
        ax.invert_yaxis()
        ax.axis("off")

    # 删除多余子图
    for j in range(len(genes), len(axes)):
        axes[j].axis("off")

    # 公共 colorbar
    cbar_ax = fig.add_axes([0.88, 0.2, 0.02, 0.6])  # [left, bottom, width, height]
    cbar = fig.colorbar(sc, cax=cbar_ax)
    cbar.set_label(f"{'weighted density' if mode=='weighted' else 'density'}")

    if save is not None:
        plt.savefig(save, format="png", dpi=300, bbox_inches="tight")
        print(f"✅ 多基因密度图已保存: {save}")
    else:
        plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

def plot_avg_gene_density(
    adata, 
    genes, 
    score_key="score_mtx", 
    tumor_label="Tumor cells",
    mode="weighted",   # "density"=普通密度, "weighted"=表达量加权密度
    point_size=8,
    cmap="viridis",
    figsize=(6,6),
    bg_size=5,
    bg_alpha=0.3,
    save=None
):
    """
    绘制多个基因平均表达的密度点图
    - tumor_label 对应的细胞参与 KDE 并着色
    - 其他细胞标灰色背景
    """

    if score_key not in adata.obsm:
        raise KeyError(f"{score_key} not found in adata.obsm")

    score_mtx = adata.obsm[score_key]

    # 检查基因是否存在
    genes_exist = [g for g in genes if g in score_mtx.columns]
    if len(genes_exist) == 0:
        raise ValueError("None of the provided genes exist in score_key matrix")

    # 计算每个细胞的平均表达
    avg_expr = score_mtx[genes_exist].mean(axis=1).values

    # 背景数据
    mask_tumor = adata.obs["celltype"] == tumor_label
    mask_other = ~mask_tumor
    coords_other = np.asarray(adata.obsm["spatial"][mask_other], dtype=float)

    # 肿瘤细胞数据
    coords_tumor = np.asarray(adata.obsm["spatial"][mask_tumor], dtype=float)
    expr_tumor = avg_expr[mask_tumor]

    # KDE 密度计算
    kde = gaussian_kde(coords_tumor.T)
    densities = kde(coords_tumor.T)

    if mode == "weighted":
        expr_shifted = expr_tumor - np.nanmin(expr_tumor)
        densities = densities * expr_shifted

    # 排序避免高值被遮挡
    idx = densities.argsort()
    coords_tumor = coords_tumor[idx]
    densities = densities[idx]

    # 绘图
    plt.figure(figsize=figsize)

    # 背景
    plt.scatter(
        coords_other[:,0],
        coords_other[:,1],
        c="lightgray",
        s=bg_size,
        alpha=bg_alpha,
        edgecolors="none"
    )

    # 肿瘤细胞
    sc = plt.scatter(
        coords_tumor[:,0],
        coords_tumor[:,1],
        c=densities,
        s=point_size,
        cmap=cmap,
        alpha=0.9,
        edgecolors="none"
    )

    plt.gca().invert_yaxis()
    plt.axis("off")
    cbar = plt.colorbar(sc)
    cbar.set_label(f"{'weighted density' if mode=='weighted' else 'density'}")

    if save:
        plt.savefig(save, format="png", dpi=300, bbox_inches="tight")
        print(f"✅ 多基因平均密度图已保存: {save}")
    else:
        plt.show()
```

```python
def plot_multi_gene_expression(
    adata, 
    genes, 
    score_key="score_mtx", 
    tumor_label="Tumor cells",
    point_size=8,
    cmap="viridis",
    ncols=2,
    figsize=(12, 5),
    bg_size=5,
    bg_alpha=0.3,
    share_scale=False,
    save=None
):
    """
    绘制多个基因的表达值点图 (0–1归一化)
    """

    if score_key not in adata.obsm:
        raise KeyError(f"{score_key} not found in adata.obsm")

    score_mtx = adata.obsm[score_key]

    n_genes = len(genes)
    nrows = int(np.ceil(n_genes / ncols))

    fig, axes = plt.subplots(
        nrows, ncols, 
        figsize=(figsize[0], figsize[1] * nrows),
        gridspec_kw={"wspace": 0.2, "hspace": 0.3, "right": 0.85}
    )
    axes = np.array(axes).reshape(-1)

    mask_tumor = adata.obs["celltype"] == tumor_label
    mask_other = ~mask_tumor

    coords_other = np.asarray(adata.obsm["spatial"][mask_other], dtype=float)

    all_values = []
    expr_dict = {}

    for gene in genes:
        if gene not in score_mtx.columns:
            continue

        coords_tumor = np.asarray(adata.obsm["spatial"][mask_tumor], dtype=float)
        expr_tumor = np.asarray(
            score_mtx.loc[adata.obs_names[mask_tumor], gene].values,
            dtype=float
        )

        # ✅ 关键：scale到0-1
        min_val = np.nanmin(expr_tumor)
        max_val = np.nanmax(expr_tumor)
        if max_val > min_val:
            expr_tumor = (expr_tumor - min_val) / (max_val - min_val)
        else:
            expr_tumor = np.zeros_like(expr_tumor)

        expr_dict[gene] = (coords_tumor, expr_tumor)
        all_values.append(expr_tumor)

    vmin = min(np.min(v) for v in all_values) if share_scale else None
    vmax = max(np.max(v) for v in all_values) if share_scale else None

    sc = None
    for i, gene in enumerate(genes):
        ax = axes[i]

        if gene not in expr_dict:
            ax.set_title(f"{gene} (not found)")
            ax.axis("off")
            continue

        coords_tumor, expr_tumor = expr_dict[gene]

        idx = expr_tumor.argsort()
        coords_tumor = coords_tumor[idx]
        expr_tumor = expr_tumor[idx]

        ax.scatter(
            coords_other[:, 0],
            coords_other[:, 1],
            c="lightgray",
            s=bg_size,
            alpha=bg_alpha,
            edgecolors="none"
        )

        sc = ax.scatter(
            coords_tumor[:, 0],
            coords_tumor[:, 1],
            c=expr_tumor,
            s=point_size,
            cmap=cmap,
            alpha=0.9,
            edgecolors="none",
            vmin=vmin,
            vmax=vmax
        )

        ax.set_title(gene, fontsize=12)
        ax.invert_yaxis()
        ax.axis("off")

    for j in range(len(genes), len(axes)):
        axes[j].axis("off")

    cbar_ax = fig.add_axes([0.88, 0.2, 0.02, 0.6])
    cbar = fig.colorbar(sc, cax=cbar_ax)
    cbar.set_label("scaled expression (0–1)")

    if save is not None:
        plt.savefig(save, format="png", dpi=300, bbox_inches="tight")
        print(f"✅ 多基因表达图已保存: {save}")
    else:
        plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from sklearn.preprocessing import StandardScaler
import tqdm

def plot_samplewise_density(
    adata,
    disease_to_geneset,
    score_key="score_mtx",
    tumor_label="Tumor cells",
    scale_method="zscore",  # "zscore" or "minmax"
    point_size=8,
    cmap="viridis",
    bg_size=5,
    bg_alpha=0.3,
    save=None,
    figsize=(12, 12),
    color_range=(-2,2)
):
    """
    按 Sample 计算疾病对应的基因集密度（在肿瘤细胞内），
    每个 Sample 内部 scale，最后合并输出一张大图。

    Parameters
    ----------
    adata : AnnData
    disease_to_geneset : dict
        {Disease: geneset_name}
    score_key : str
        adata.obsm 中的评分矩阵键
    tumor_label : str
        肿瘤细胞标签
    scale_method : str
        "zscore" (StandardScaler) 或 "minmax"
    """
    if score_key not in adata.obsm:
        raise KeyError(f"{score_key} not in adata.obsm")

    score_mtx = adata.obsm[score_key]
    density_all = np.full(adata.n_obs, np.nan)

    for sample in tqdm.tqdm(adata.obs["Sample"].unique(), desc="Per-sample density"):
        mask_sample = adata.obs["Sample"] == sample
        disease = adata.obs.loc[mask_sample, "Disease"].unique()[0]
        geneset = disease_to_geneset.get(disease, None)

        if geneset is None or geneset not in score_mtx.columns:
            continue

        mask_tumor = mask_sample & (adata.obs["celltype"] == tumor_label)
        if np.sum(mask_tumor) == 0:
            continue

        coords_tumor = np.asarray(adata.obsm["spatial"][mask_tumor], dtype=float)
        expr_tumor = np.asarray(score_mtx.loc[adata.obs_names[mask_tumor], geneset].values, dtype=float)

        # KDE
        kde = gaussian_kde(coords_tumor.T)
        densities = kde(coords_tumor.T)

        # 加权 (density * expr)
        expr_shifted = expr_tumor - np.nanmin(expr_tumor)
        densities = densities * expr_shifted

        # 在 sample 内部做 scale
        if scale_method == "zscore":
            densities_scaled = StandardScaler().fit_transform(densities.reshape(-1, 1)).flatten()
        elif scale_method == "minmax":
            vmin, vmax = np.min(densities), np.max(densities)
            densities_scaled = (densities - vmin) / (vmax - vmin + 1e-9)
        else:
            raise ValueError("scale_method must be 'zscore' or 'minmax'")

        density_all[mask_tumor] = densities_scaled

    # 保存结果
    adata.obs["sample_scaled_density"] = density_all

    # 绘制
    mask_tumor_all = adata.obs["celltype"] == tumor_label
    mask_other = ~mask_tumor_all

    coords_other = np.asarray(adata.obsm["spatial"][mask_other], dtype=float)
    coords_tumor = np.asarray(adata.obsm["spatial"][mask_tumor_all], dtype=float)
    density_vals = density_all[mask_tumor_all]

    plt.figure(figsize=figsize)
    plt.scatter(coords_other[:, 0], coords_other[:, 1],
                c="lightgray", s=bg_size, alpha=bg_alpha, edgecolors="none")
    sc = plt.scatter(coords_tumor[:, 0], coords_tumor[:, 1],
                     c=density_vals, s=point_size, cmap=cmap,
                     alpha=0.9, edgecolors="none",vmax=color_range[1])
    plt.gca().invert_yaxis()
    plt.axis("off")
    plt.colorbar(sc, label="scaled density")

    if save is not None:
        plt.savefig(save, dpi=300, bbox_inches="tight")
        print(f"✅ 图已保存: {save}")
    else:
        plt.show()

    return adata
```

```python
def plot_cnv_spatial(
    adata,
    features,
    obsm_key_cnv="cnv_mtx",
    obsm_key_spatial="spatial",
    vmin=None,
    vmax=None,
    cmap="coolwarm",
    point_size=10,
    ncols=2,
    figsize=(12, 12)   # <-- 增加 figsize 参数
):
    coords = adata.obsm[obsm_key_spatial]
    cnv_data = adata.obsm[obsm_key_cnv]

    if not isinstance(cnv_data, pd.DataFrame):
        cnv_data = pd.DataFrame(
            cnv_data,
            index=adata.obs_names,
            columns=[f"CNV_{i}" for i in range(cnv_data.shape[1])]
        )

    missing = [f for f in features if f not in cnv_data.columns]
    if missing:
        raise ValueError(f"以下 features 不在 {obsm_key_cnv} 中: {missing}")

    nrows = int(np.ceil(len(features) / ncols))
    fig, axes = plt.subplots(
        nrows, ncols,
        figsize=figsize,  # 使用 figsize 参数控制整体大小
        constrained_layout=True
    )

    if nrows == 1 and ncols == 1:
        axes = np.array([[axes]])
    elif nrows == 1 or ncols == 1:
        axes = axes.reshape(nrows, ncols)

    for ax, feature in zip(axes.flatten(), features):
        values = cnv_data[feature].values
        sc = ax.scatter(
            coords[:,0], coords[:,1], c=values,
            cmap=cmap, s=point_size, alpha=0.9,
            vmin=(np.min(values) if vmin is None else vmin),
            vmax=(np.max(values) if vmax is None else vmax)
        )
        ax.invert_yaxis()
        ax.set_aspect("equal")  # 保持空间比例
        ax.axis("off")
        ax.set_title(feature)
        fig.colorbar(sc, ax=ax, fraction=0.046, pad=0.04)

    for ax in axes.flatten()[len(features):]:
        ax.axis("off")

    plt.show()
```
