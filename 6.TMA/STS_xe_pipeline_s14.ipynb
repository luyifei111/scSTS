```python
adata_TMA_merged_2 = sc.read_h5ad("/cluster3/yflu/STS/TMA/adata_TMA_merged.h5ad")
```

```python
sample_counts = adata_TMA_merged_2.obs["Sample"].value_counts()
Samples = sample_counts[sample_counts > 0].index
Samples = Samples.tolist()
```

```python
Samples
```

```python
adata_TMA_merged_2.obs["banksy"] = "Not defined"
```

```python
for sample in Samples:
    preprocessed_path = f"/cluster3/yflu/STS/TMA/separated/{sample}/{sample}_preprocessed.h5ad"
    adata_1 = sc.read_h5ad(preprocessed_path)

    # 重新设置 index
    adata_1.obs.index = adata_1.obs["sample"][0] + "-" + adata_1.obs["cell_id"].astype(str)

    # 如果没有 banksy_domain 列，跳过该样本
    if "banksy_domain" not in adata_1.obs.columns:
        print(f"Skipping {sample}, no banksy_domain in obs.")
        continue

    # 构建新 banksy 列
    adata_1_new_banksy = sample + "_" + adata_1.obs["banksy_domain"].astype(str)

    # 替换对应索引的值
    adata_TMA_merged_2.obs.loc[adata_1.obs.index, "banksy"] = adata_1_new_banksy
    
    print(f"{sample}, done.")
```
